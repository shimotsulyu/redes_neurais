```python
from sklearn import linear_model, model_selection, metrics
import pandas as pd

arquivo = 'data0'
porcentagem = .8
# Carrega os elementos do dataset
dataset = pd.read_csv(f'data/simples/{arquivo}.csv')
x = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values

# Cria um objeto Perceptron
perceptron = linear_model.Perceptron()

# Split em conjunto de treino e teste
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=1-porcentagem, random_state=0)

# Treinamento
classificador = perceptron.fit(x_train, y_train)

# Validação
y_predict = classificador.predict(x_test)

# Acurácia
print(f'Precisão: {metrics.accuracy_score(y_test, y_predict)*100}%')


dataset.plot.scatter(x=1,y=2, #x,y
                    c=0, #classes
                    colormap='jet')

x_test = pd.DataFrame(x_test,columns=[[1,2]])
y_predict = pd.DataFrame(y_predict)
x_test[0] = y_predict
x_test.plot.scatter(x=0,y=1, #x,y
                    c=2, #classes
                    colormap='jet')
```
